```python
from sklearn.preprocessing import LabelEncoder
```

```python
dfle = pd.read_csv('carprices.csv')
dfle
```

```python
le = LabelEncoder()
```

```python
dfle['Car Model'] = le.fit_transform(dfle['Car Model'])
```

```python
dfle
```

```python
ohe = OneHotEncoder()
```

```python
test= ohe.fit_transform(dfle[['Car Model']]).toarray()
test
```

```python
new_test_df = pd.DataFrame(test, columns=['category_0', 'category_1', 'category_2'])
```

```python
new_test_df
```

```python
final_df = pd.concat([dfle, new_test_df], axis = 1)
```

```python
final_df
```

```python
X = final_df.drop(['Car Model', 'category_0', 'Sell Price($)'], axis = 1)
X
```

```python
y = final_df['Sell Price($)']
y
```

```python
lr = LinearRegression()
```

```python
lr.fit(X, y)
```
